### 4.2 Analyzing *UI_EVENT_SOURCE*

```python
lightning_performance_simplified['UI_EVENT_SOURCE'].unique()
```

Numerical data have been identified

#### 4.2.1 Wiping the data

```python
# Getting the digit components
digit_components = []
for component in lightning_performance_simplified['UI_EVENT_SOURCE'].unique():
    if component[0].isdigit():
        digit_components.append(component)
```

```python
clean_data = lightning_performance_simplified
for component in digit_components:
    dirty_data_idx = lightning_performance_simplified[lightning_performance_simplified['UI_EVENT_SOURCE'] == component].index
    clean_data = clean_data.drop(dirty_data_idx)
```

```python
# Checking the data
clean_data['UI_EVENT_SOURCE'].unique()
```

#### 4.2.2 Couting components

```python
from collections import defaultdict

components_count = defaultdict(int)

for component in clean_data['UI_EVENT_SOURCE']:
    components_count[str(component)] += 1
```

```python
components_count
```

```python
# Creating a DataFrame with the dictionary
components_df = pd.DataFrame(components_count.items(), columns=['Component', 'Count'])
```

```python
# Sorting by count
components_df_sorted = components_df.sort_values(by=['Count'], ascending=False)
```

```python
import matplotlib.pyplot as plt

number_of_components = 10

components_df_sorted.head(number_of_components).plot(kind='bar', x="Component")
```
