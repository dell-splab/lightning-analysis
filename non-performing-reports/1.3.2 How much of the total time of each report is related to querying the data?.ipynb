```python
run_report_api_resource = "/lightning/r/"

def get_report_info_from_endpoint(page_url):
    if page_url.startswith(run_report_api_resource):
        uri = page_url.split(run_report_api_resource)
        if len(uri) >= 2:
            report_type, report_id_derived = uri[1].split("/")[:2]
            return report_type, report_id_derived
```

```python
active_and_problematics['REPORT_INFO_DERIVED'] = active_and_problematics.PageUrl.apply(get_report_info_from_endpoint)
```

```python
active_and_problematics.dropna(subset=['REPORT_INFO_DERIVED'], inplace=True)
```

```python
active_and_problematics['REPORT_TYPE_DERIVED'] = active_and_problematics.REPORT_INFO_DERIVED.str[0]
active_and_problematics['REPORT_ID_DERIVED'] = active_and_problematics.REPORT_INFO_DERIVED.str[1]
```
